# Cây quyết định
## 1. Mô hình
- **Cây quyết định (Decision Tree)** là một thuật toán học máy có giám sát, được sử dụng trong các bài toán phân loại và hồi quy. Về mặt trực quan, nó được biểu diễn dưới dạng một sơ đồ nhánh, trong đó mỗi nút đại diện cho một điều kiện, quy tắc, kết quả hoặc nhóm phân loại.
- **Nguyên lý phân chia dữ liệu:** Tại mỗi nút, mô hình sẽ duyệt qua các đặc trưng và thử các ngưỡng chia khác nhau. Mục tiêu là tìm ra cách chia giúp dữ liệu ở các nút con trở nên đồng nhất nhất có thể - nghĩa là mỗi nút con chứa càng nhiều mẫu cùng một loại càng tốt.

 

## 2. Bài toán áp dụng
- Áp dụng cho cả bai toán **Phân loại** và **Hồi quy**
- **Cây phân loại**: dự báo giá trị biến phụ thuộc bằng giá trị của biến có tỷ lệ cao nhất trong node lá
- **Cây hồi quy**: dự báo giá trị biến phụ thuộc bằng trung bình giá trị các biến phụ thuộc trong node lá

## 3. Thuật toán xây dựng cây
### 3.1 **ID3**
- Thuật toán này sử dụng khái niệm **entropy** để đo độ hỗn loạn, xáo trộn của các lớp trong tập dữ liệu
- Mục tiêu của **ID3** là tìm ra thuộc tính mang lại lượng thông tin thu được **(IG)** lớn nhất để làm node chia nhánh
- Tập $S$ có số phần tử $|S| = N$, trong đó có C lớp, số phần tử của mỗi tập là $N_i$
- Entropy của tập trước khi chia:
  $$H(S) = - \sum_{i=1}^{C}\frac{N_i}{N}log\left(\frac{N_i}{N}\right)$$
- Giả sử thuộc tính được chọn để phân chia tập là $x$. Dựa trên $x$, dữ liệu trên $S$ được chia thành $k$ tập con $S_1, S_2,.., S_k$ với số phần tử là $m_1, m_2, .., m_k$
- Entropy sau khi chia theo thuộc tính $x$:
$$H(S, x) = \sum_{i=1}^{k}\frac{m_i}{N}H(S_k)$$
$$IG(S, x) = H(S) - H(S, x)$$
- Trong **ID3** thuộc tính $x$ được chọn sao cho lượng thông tin thu được(IG) lớn nhất
$$x^* = \arg \max_x IG(\mathcal{S}, x) = \arg \min_x H(\mathcal{S}, x)$$

  - Luôn chọn các thuộc tính có nhiều giá trị khác nhau tạo ra cây có kích thước lớn, dẫn đến hiện tượng quá khớp

  - Dễ bị học vẹt do phát triển cây quá sâu cho đến khi node lá thuần nhất

### 3.2. C4.5
- Cách xử lý của **C4.5**:
  - Đưa ra hệ số phạt Split Information:
    $$SplitInfo(S, A) = - \sum_{v \in Values(A)} \frac{\vert{}S_v\vert{}}{\vert{}S\vert{}} \log_2\left(\frac{\vert{}S_v\vert{}}{\vert{}S\vert{}}\right)$$
  - Tỷ lệ thông tin thu được:
    $$GainRatio(S, A) = \frac{IG(S, A)}{SplitInfo(S, A)}$$
    Thuật toán sẽ chọn thuộc tính có Gain Ratio lớn nhất để làm node phân chia, thay vì IG lớn nhất.
- Cơ chế xử lý dữ liệu liên tục: sắp xếp mảng và tính các điểm ngưỡng nằm giữa 2 giá trị liền kề. Tính Gain Ratio cho tất cả các điểm ngưỡng và chọn điểm cắt tối ưu nhất để biến thuộc tính liên tục thành thuộc tính nhị phân
- Xử lý giá trị bị thiếu: khi gặp giá trị bị thiếu, tạm bỏ qua dòng đó để tính Gain Ratio, khi quyết định chia dữ liệu xuống các nhánh con, giá trị còn thiếu sẽ được đẩy vào tất cả các nhánh với tỉ lệ dữ liệu mỗi nhánh này với tổng dữ liệu
- Chống quá khớp: xây dựng cây đến tận cùng sau đó cắt tỉa cây. Đi ngược từ lá ước lượng tỉ lệ lỗi của các nhánh con, nếu việc gộp nhánh giúp giảm tỉ lệ lỗi nó sẽ gộp nhành và cắt bỏ nhành thừa

Ví dụ:

| Mã SV | Điểm Giữa Kỳ | Chuyên Cần | Bài Tập Lớn | Qua Môn|
| --- | --- | --- | --- | --- |
| SV01 | 8.5 | Có | Tốt | **Đỗ** |
| SV02 | 4.0 | Không | Kém | **Trượt** |
| SV03 | 6.5 | Có | Trung bình | **Đỗ** |
| SV04 | 5.0 | Không | Kém | **Trượt** |
| SV05 | 7.0 | *Trống* | Tốt | **Đỗ** |
| SV06 | 9.0 | Có | Tốt | **Đỗ** |
| SV07 | 3.5 | Không | Trung bình | **Trượt** |
| SV08 | 5.5 | *Trống* | Kém | **Trượt** |
| SV09 | 8.0 | Có | Trung bình | **Đỗ** |
| SV10 | 4.5 | Có | Tốt | **Trượt** |

$$H(S) = -\frac{5}{10}\log_2\left(\frac{5}{10}\right) - \frac{5}{10}\log_2\left(\frac{5}{10}\right) = \mathbf{1.0}$$

- Chọn thuộc tính **mã SV**
  $$IG = H(S) - H(S, MSV) = 1.0$$
  $$SplitInfo(S, MSV) = -10\frac{1}{10}log_2\left(\frac{1}{10}\right)=\log_2(10) \approx 3.32$$
  $$GainRatio(S, MSV)=\frac{1.0}{3.32}=0.30$$
- Chọn thuộc tính **Điểm giữa kỳ**
  - Sau khi sắp xếp: $3.5$ (Trượt), $4.0$ (Trượt), $4.5$ (Trượt), $5.0$ (Trượt), $5.5$ (Trượt) | $6.5$ (Đỗ), $7.0$ (Đỗ), $8.0$ (Đỗ), $8.5$ (Đỗ), $9.0$ (Đỗ)
  - Thấy điểm cắt tối ưu là $\frac{5.5 + 6.5}{2} = 6.0$
  - GainRatio tại $6.0$:
    $$IG(S, \text{Điểm}) = H(S) - \left( \frac{5}{10}H(S_{<6.0}) + \frac{5}{10}H(S_{\ge 6.0}) \right)$$
    $$IG(S, \text{Điểm}) = 1.0 - \left( \frac{5}{10} \times 0 + \frac{5}{10} \times 0 \right) = 1.0$$
    $$SplitInfo(S, \text{Điểm}) = - \left( \frac{5}{10}\log_2\left(\frac{5}{10}\right) + \frac{5}{10}\log_2\left(\frac{5}{10}\right) \right) = 1.0$$
    $$GainRatio(S, \text{Điểm}) = \frac{1.0}{1.0} = 1.0$$
- Chọn thuộc tính **Chuyên cần**

| Mã SV | Điểm Giữa Kỳ | Chuyên Cần | Bài Tập Lớn | Qua Môn|
| --- | --- | --- | --- | --- |
| SV01 | 8.5 | Có | Tốt | **Đỗ** |
| SV03 | 6.5 | Có | Trung bình | **Đỗ** |
| SV06 | 9.0 | Có | Tốt | **Đỗ** |
| SV09 | 8.0 | Có | Trung bình | **Đỗ** |
| SV10 | 4.5 | Có | Tốt | **Trượt** |

| Mã SV | Điểm Giữa Kỳ | Chuyên Cần | Bài Tập Lớn | Qua Môn|
| --- | --- | --- | --- | --- |
| SV02 | 4.0 | Không | Kém | **Trượt** |
| SV04 | 5.0 | Không | Kém | **Trượt** |
| SV07 | 3.5 | Không | Trung bình | **Trượt** |

| Mã SV | Điểm Giữa Kỳ | Chuyên Cần | Bài Tập Lớn | Qua Môn|
| --- | --- | --- | --- | --- |
| SV05 | 7.0 | *Trống* | Tốt | **Đỗ** |
| SV08 | 5.5 | *Trống* | Kém | **Trượt** |

$$H(S, chuyen\_can) = -\frac{5}{8}\left( \frac{4}{5}log_2\frac{4}{5} +  \frac{1}{5}log_2\frac{1}{5} \right)-\frac{3}{8}\frac{3}{3}log_2\frac{3}{3}=0.45$$
$$IG(S, chuyen\_can) = \frac{8}{10}(H(S) - H(S, chuyen\_can)) = 0.8 (1.0 - 0.45) = 0.44 $$
$$SplitInfo = -\left( \frac{5}{10}\log_2\left(\frac{5}{10}\right) + \frac{3}{10}\log_2\left(\frac{3}{10}\right) + \frac{2}{10}\log_2\left(\frac{2}{10}\right) \right)=1.485$$
$$GainRatio = \frac{IG}{SplitInfo} = \frac{0.44}{1.485} \approx \mathbf{0.30}$$
- Chọn thuộc tính **Bài tập lớn**
  
| Mã SV | Điểm Giữa Kỳ | Chuyên Cần | Bài Tập Lớn | Qua Môn|
| --- | --- | --- | --- | --- |

| --- | --- | --- | --- | --- |

$$H(S, chuyen\_can) = -\frac{4}{10}\left( \frac{3}{4}log_2\frac{3}{4} + \frac{1}{4}log_2\frac{1}{4} \right)
                    -\frac{3}{10}\cdot 0
                    -\frac{3}{10}\left( \frac{2}{3}log_2\frac{2}{3} + \frac{1}{3}log_2\frac{1}{3} \right) = 0.6$$
$$IG(S, chuyen\_can) = 1.0 - 0.6 = 0.4$$
$$SplitInfo = -\frac{4}{10}log_2\frac{4}{10}-\frac{3}{10}log_2\frac{3}{10}-\frac{3}{10}log_2\frac{3}{10} = 1.57$$
$$GainRatio = \frac{0.4}{1.57} = 0.25$$

### 3.3. CART
- CART giải quyết cả bài toán Phân loại và Hồi quy, là thuật toán nền tảng và phổ biến trong ML
- Khác với ID3 hay C4.5 nó chỉ cho phép tại mỗi Nút chỉ có 2 nhánh con YES/NO

- **Ý tưởng cốt lõi:** chia để trị bằng các luật ngưỡng

Có tập dữ liệu gồm $n$ mẫu, mỗi mẫu là một cặp (đặc trưng, nhãn). CART xây cây bằng cách **chia nhị phân lặp đi lặp lại** — gọi là *đệ quy* (đơn giản là: chia xong một node thì áp dụng đúng quy trình đó cho hai node con).

Tại mỗi node đang xét, thuật toán chọn ra **một đặc trưng $j$** (ví dụ: cột "tuổi") và **một ngưỡng $t$** (ví dụ: 30), rồi chia dữ liệu tại node đó thành 2 nhóm:

- **Nhóm trái:** các mẫu có giá trị đặc trưng $j$ $\le t$ (ví dụ: tuổi $\le 30$)
- **Nhóm phải:** các mẫu có giá trị đặc trưng $j$ $> t$ (ví dụ: tuổi $> 30$)

Việc chia này biến không gian đặc trưng thành các **vùng chữ nhật** — cây càng sâu, các vùng càng nhỏ và "thuần khiết" hơn (các điểm trong cùng một vùng có nhãn/giá trị càng giống nhau).

Câu hỏi còn lại: **làm sao đo "một cách chia tốt"?** Đó là lý do ta cần một hàm **impurity**.


### Impurity
**Với bài toán phân loại**:
- **Gini Index (mặc định trong CART):**
$$\text{Gini} = 1 - \sum_{k=1}^{K} p_k^2$$
Với bài toán 2 lớp (tỷ lệ lớp 1 là p, lớp 2 là 1-p):
$$Gini = 1-(1-p)^2 - p^2 = 2p(1-p)$$
Đạt giá trị lớn nhất $0.5$ khi $p = 0.5$ (hai lớp bằng nhau), và bằng $0$ khi $p = 0$ hoặc $p = 1$ (node thuần khiết).

- **Entropy:**
$$\text{Entropy} = -\sum_{k=1}^{K} p_k \log_2 p_k$$
Với 2 lớp: $\text{Entropy} = -p\log_2 p - (1-p)\log_2(1-p)$, cũng đạt lớn nhất (bằng $1$) khi $p=0.5$.

- **Misclassification Error (lỗi phân loại):**

$$\text{Error} = 1 - \max_k p_k$$

(1 trừ tỉ lệ của lớp đông nhất). Ít được dùng để **chọn điểm chia** vì kém nhạy hơn Gini/Entropy — đồ thị của nó có "góc nhọn" nên khó phân biệt các điểm chia gần nhau — nhưng vẫn hữu ích để **đánh giá và cắt tỉa** cây sau khi xây xong.
**So sánh trực quan ba độ đo (bài toán 2 lớp)**

````python
import numpy as np
import matplotlib.pyplot as plt

p = np.linspace(1e-6, 1 - 1e-6, 400)

gini = 2 * p * (1-p)
entropy = -p*np.log2(p)-(1-p)*np.log2(1-p)
missclass_error = 1 - np.maximum(p, 1-p)

fig, ax = plt.subplots(figsize=(7.5, 5))

ax.plot(p, gini, label='Gini Index: $2p(1-p)$', lw=2.5, color='#2563eb')
ax.plot(p, entropy, label='Entropy: $-p\\log_2 p-(1-p)\\log_2(1-p)$', lw=2.5, color='#dc2626')
ax.plot(p, missclass_error, label='Misclassification Error: $1-\\max(p,1-p)$', lw=2.5, color='#16a34a')

ax.set_xlabel('$p$ = tỉ lệ lớp dương trong node')
ax.set_ylabel('Giá trị impurity')
ax.set_title('So sánh các độ đo Impurity cho bài toán 2 lớp')
ax.axvline(0.5, color='gray', ls=':', lw=1)
ax.legend(loc='upper center', fontsize=9.5)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('impurity_comparison.png', bbox_inches='tight')
plt.show()
````

**Với bài toán hồi quy:**
- Nhãn y là số thực, Impurity của một node được định nghĩa là **sai số bình phương trung bình (MSE)** so với giá trị dự đoán tốt nhất có thể (là trung bình cộng các y trong node)
$$\text{MSE} = \frac{1}{n}\sum_{i=1}^{n} (y_i - \bar{y})^2, \qquad \bar{y} = \text{trung bình cộng các } y_i \text{ trong node}$$
- **Vì sao trung bình là dự đoán tối ưu?** Nếu phải chọn một con số $c$ duy nhất để dự đoán cho cả node sao cho tổng bình phương sai số $\sum_i (y_i - c)^2$ nhỏ nhất, thì $c$ tốt nhất chính là trung bình cộng $\bar y$

### Tiêu chí phân chia
Giả sử node đang xét có $n$ mẫu, impurity là $\text{Imp}$. Khi ta thử chia theo một cặp (đặc trưng, ngưỡng) nào đó, node tách thành:

- Nhóm trái: $n_L$ mẫu, impurity $\text{Imp}_L$
- Nhóm phải: $n_R$ mẫu, impurity $\text{Imp}_R$ (với $n_L + n_R = n$)

**Impurity trung bình sau khi chia** (lấy trọng số theo kích thước từng nhóm, vì nhóm to hơn thì "đáng tin" hơn):

$$\text{Imp}_{\text{sau chia}} = \frac{n_L}{n}\,\text{Imp}_L + \frac{n_R}{n}\,\text{Imp}_R$$

**Information Gain** = impurity giảm được so với trước khi chia:

$$\text{Gain} = \text{Imp} - \text{Imp}_{\text{sau chia}}$$

CART thử **tất cả** các cặp (đặc trưng, ngưỡng) có thể, rồi **chọn cách chia cho Gain lớn nhất** (tương đương: chọn cách chia cho $\text{Imp}_{\text{sau chia}}$ nhỏ nhất, vì $\text{Imp}$ ban đầu là số cố định, không đổi theo cách chia).


> **Lưu ý quan trọng — tính tham lam:** CART chọn cách chia tốt nhất cho **từng node một**, không nhìn xa hơn để tối ưu toàn bộ cây cùng lúc — gọi là chiến lược "tham lam" (greedy). Cách này không đảm bảo cây thu được là tối ưu tuyệt đối trên toàn cây (tìm cây tối ưu tuyệt đối là bài toán cực khó về mặt tính toán), nhưng đổi lại tốc độ xây cây rất nhanh.

### Các điều kiện dừng thường dùng

| Tham số | Ý nghĩa |
|---|---|
| `max_depth` | Độ sâu tối đa của cây |
| Số mẫu tối thiểu để tiếp tục chia | Node có ít hơn ngưỡng này thì không chia nữa |
| Số mẫu tối thiểu ở mỗi leaf | Không cho phép leaf quá nhỏ |
| Node đã thuần khiết (impurity = 0) | Không còn gì để cải thiện |
| Gain không dương / dưới ngưỡng | Chia không giúp ích |
| Số lá tối đa | Giới hạn tổng kích thước cây |

###  Độ quan trọng của đặc trưng (Feature Importance)
Ý tưởng rất đơn giản: một đặc trưng được coi là "quan trọng" nếu nó giúp làm giảm impurity nhiều mỗi khi được chọn để chia node, đặc biệt là ở những node có nhiều mẫu (vì ảnh hưởng tới nhiều dữ liệu hơn).

$$\text{importance}(feature\ j) = \sum_{\text{node } t \text{ chia theo } j} \frac{n_t}{n_{total}} \times \left(\text{impurity trước khi chia} - \text{impurity trung bình sau khi chia}\right)$$

trong đó $n_t$ là số mẫu tại node $t$, $n_{total}$ là tổng số mẫu train. Sau đó chuẩn hoá để tổng các importance của mọi đặc trưng bằng 1 (để dễ đọc theo %).


### Cắt tỉa cây (Tree pruning)
- Như ở trên ta đã chặn không cho cây phát triển bằng cách dừng sớm: `max_depth`, `min_samples_split`, `min_samples_leaf`,...Cách này gọi là **pre-pruning** (cắt tỉa trước) ngăn chặn cây mọc quá lớn ngay từ đầu
- Một cách khác là **post-pruning** (cắt tỉa sau): cho cây mọc **hết cỡ** trước (rất dễ overfitting), sau đó **cắt bớt** những nhánh không đóng góp nhiều, dựa trên một quy tắc gọi là **Cost-Complexity Pruning**.
- **Ý tưởng công thức**: với mỗi cây con $T$ cắt ra từ cây gốc, ta định nghĩa một "chi phí" gồm 2 phần — sai số của cây và độ phức tạp (số lá) của cây:
$$R_\alpha(T) = R(T) + \alpha \times |T|$$

    - $R(T)$: tổng impurity (hoặc sai số) của cây $T$ trên toàn bộ dữ liệu train.
    - $|T|$: số node lá của cây $T$ — cây càng nhiều lá càng "phức tạp", càng dễ overfitting.
    - $\alpha \ge 0$: hệ số phạt độ phức tạp, do người dùng chọn (giống vai trò $\lambda$ trong Ridge/Lasso mà bạn đã học).

Hiểu đơn giản: $\alpha$ càng lớn, ta càng "phạt nặng" các cây có nhiều lá, khiến thuật toán ưu tiên cắt bớt các nhánh chỉ giúp giảm sai số rất ít nhưng lại làm cây phình to. $\alpha = 0$ tức là không phạt gì cả — giữ nguyên cây đầy đủ (dễ overfitting nhất).

````python
from typing import Any

class Node:
    __slots__ = ('feature', 'threshold', 'left', 'right', 'depth', 'value', 'n_samples', 'impurity')

    def __init__(self, depth: int):
            self.feature: int | None = None
            self.threshold: float | None= None
            self.left: Node | None = None
            self.right: Node | None = None
            self.depth: int = depth
            self.value: Any = None
            self.n_samples: int | None = None
            self.impurity: float | None = None

    def is_leaf(self):
        return self.value is not None
````

````python
import numpy as np
from Node import Node
from typing import Dict

def gini(counts):
    n = np.sum(counts)
    if n == 0:
        return 0
    p = counts / n
    return 1 - np.sum(p ** 2)

def entropy(counts):
    n = np.sum(counts)
    if n == 0:
        return 0
    p = counts / n
    p = p[p > 0]
    log_p = np.log2(p)
    return -np.sum(p  * log_p)

IMPURITY_FUNC = {'gini': gini, 'entropy': entropy}

class CARTClassifier:
    def __init__(self, criterion='gini', max_depth = None, max_features = None, min_samples_split = 2, min_samples_leaf = 1, ccp_alpha = 0, min_impurity_decrease = 0):
        self.criterion = criterion
        self.max_depth = max_depth
        self.max_features = max_features
        self.min_samples_split = min_samples_split
        self.min_samples_leaf = min_samples_leaf
        self.min_impurity_decrease = min_impurity_decrease
        self.root = None
        self.n_classes_ = 0
        self.n_leaves_ = 0
        self._impurity_func = IMPURITY_FUNC[criterion]
        self.n_total_samples = 0
        self.feature_importances = None
        self.ccp_alpha = ccp_alpha

    def fit(self, X, y):
        X = np.array(X, dtype=float)
        y = np.array(y, dtype=int)
        n_samples, n_features  = X.shape
        self.n_total_samples = n_samples
        self.feature_importances = np.zeros(n_features)
        self.n_classes_ = int(max(y)) + 1
        self.n_leaves_ = 0
        self.root = self._build(X, y, depth = 0)
        total = self.feature_importances.sum()
        if total > 0:
            self.feature_importances /= total
        return self

    def predict_proba(self, X):
        X = np.asarray(X, dtype=float)
        return np.array([self._predict_one(x, self.root) for x in X])

    def predict(self, X):
        return np.argmax(self.predict_proba(X), axis = 1)

    def _build(self, X, y, depth):
        node = Node(depth)
        n_samples = len(y)
        node.n_samples = n_samples
        counts = np.bincount(y, minlength=self.n_classes_)
        node.impurity = self._impurity_func(counts)
        stop = (
            node.impurity == 0 or
            n_samples < self.min_samples_split or
            (self.max_depth is not None and depth >= self.max_depth)
        )
        split: Dict | None = None
        if not stop:
            split = self. _best_split(X, y, counts)
            if split is None or split['gain'] <= self.min_impurity_decrease:
                stop = True

        if stop or split is None:
            node.value = counts / n_samples
            self.n_leaves_ += 1
            return node
        if self.feature_importances is not None:
            self.feature_importances[split['feature']] += (n_samples / self.n_total_samples) * split['gain']

        j, t = split['feature'], split['threshold']

        left = X[:, j] <= t
        node.feature = j
        node.threshold = t
        node.left = self._build(X[left], y[left], depth+1)
        node.right = self._build(X[~left], y[~left], depth+1)
        return node

    def _best_split(self, X, y, parent_counts):
        n_samples, n_features = X.shape
        parent_impurity = self._impurity_func(parent_counts)
        best = None

        if self.max_features is None:
            n_sub_features = n_features
        elif self.max_features == 'sqrt':
            n_sub_features = int(np.sqrt(n_features))
        elif self.max_features == 'third':
            n_sub_features = max(1, int(n_features / 3))
        elif isinstance(self.max_features, int):
            n_sub_features = min(self.max_features, n_features)
        elif isinstance(self.max_features, float):
            n_sub_features = max(1, int(self.max_features * n_features))
        else:
            n_sub_features = n_features

        feature_indices = np.random.choice(n_features, n_sub_features, replace=False)

        for j in feature_indices:
            col = X[:, j]
            order = np.argsort(col, kind='mergesort')
            col_sorted = col[order]
            y_sorted = y[order]

            left_counts = np.zeros(self.n_classes_)
            right_counts = parent_counts.copy()

            for i in range(n_samples - 1):
                c = y_sorted[i]
                left_counts[c] += 1
                right_counts[c] -= 1
                n_left = i+1
                n_right = n_samples - n_left

                # Nếu đặc trưng tại hai điểm có giá trị giống nhau
                if col_sorted[i] == col_sorted[i+1]:
                    continue
                # Nếu nhánh trái hoặc nhánh phải không đủ số phân tử tối thiểu
                if n_left < self.min_samples_leaf or n_right < self.min_samples_leaf:
                    continue

                imp_left = self._impurity_func(left_counts)
                imp_right = self._impurity_func(right_counts)

                weighted = (n_left / n_samples) * imp_left + (n_right / n_samples) * imp_right

                gain = parent_impurity - weighted

                if best is None or gain > best['gain']:
                    threshold = (col_sorted[i] + col_sorted[i+1]) / 2
                    best = {'feature': j, 'threshold': threshold, 'gain': gain}

        return best

    def prune(self, node):
        if node is None:
            return 0.0, 0

        if node.is_leaf:
            cost = (node.n_samples / self.n_total_samples) * node.impurity
            return cost, 1

        cost_left, leaves_left = self.prune(node.left)
        cost_right, leaves_right = self.prune(node.right)

        cost_children = cost_left + cost_right
        leaves_children = leaves_left + leaves_right

        cost_node = (node.n_samples / self.n_total_samples) * node.impurity

        if cost_node <= cost_children + self.ccp_alpha * (leaves_children - 1):
            node.left = None
            node.right = None
            self.n_leaves_ -= (leaves_children - 1)
            return cost_node, 1

        return cost_children, leaves_children

    def _predict_one(self, x, node):
        while not node.is_leaf():
            node = node.left if x[node.feature] <= node.threshold else node.right
        return node.value

    def depth(self):
        def _d(node):
            if node.is_leaf():
                return node.depth
            return max(_d(node.left), _d(node.right))
        return _d(self.root)
````

### Cài đặt CART Hồi quy

````python
from typing import Dict

import numpy as np
from Node import Node

class CARTRegressor:
    def __init__(self, max_depth = None, max_features = None, min_samples_split = 2, min_samples_leaf = 1, ccp_alpha = 0, min_mse_decrease = 0):
        self.max_depth = max_depth
        self.max_features = max_features
        self.min_samples_split = min_samples_split
        self.min_samples_leaf = min_samples_leaf
        self.min_mse_decrease = min_mse_decrease
        self.root = None
        self.n_leaves = 0
        self.n_total_samples = 0
        self.feature_importances = None
        self.ccp_alpha = ccp_alpha

    @staticmethod
    def mse(y):
        if len(y) == 0:
            return 0
        return np.var(np.asarray(y))

    def best_split(self, X, y):
        n_samples, n_features = X.shape
        S1_total, S2_total = np.sum(y), np.sum(y ** 2)
        parent_mse = self.mse(y)
        best: Dict | None = None
        if self.max_features is None:
            n_sub_features = n_features
        elif self.max_features == 'sqrt':
            n_sub_features = int(np.sqrt(n_features))
        elif self.max_features == 'third':
            n_sub_features = max(1, int(n_features / 3))
        elif isinstance(self.max_features, int):
            n_sub_features = min(self.max_features, n_features)
        elif isinstance(self.max_features, float):
            n_sub_features = max(1, int(self.max_features * n_features))
        else:
            n_sub_features = n_features

        feature_indices = np.random.choice(n_features, n_sub_features, replace=False)

        for j in feature_indices:
            col = X[:, j]
            order = np.argsort(col, kind='mergesort')
            col_sorted = col[order]
            y_sorted = y[order]

            S1_left, S2_left = 0, 0
            S1_right, S2_right = S1_total, S2_total
            for i in range(n_samples - 1):
                S1_left += y_sorted[i]
                S2_left += y_sorted[i] ** 2
                S1_right -= y_sorted[i]
                S2_right -= y_sorted[i] ** 2
                n_left = i + 1
                n_right = n_samples - n_left

                if col_sorted[i] == col_sorted[i+1]:
                    continue
                if n_left < self.min_samples_leaf or n_right < self.min_samples_leaf:
                    continue

                mse_left = S2_left / n_left - (S1_left  / n_left) ** 2
                mse_right = S2_right / n_right - (S1_right / n_right) ** 2

                weighted = (n_left / n_samples) * mse_left + (n_right / n_samples) * mse_right

                gain = parent_mse - weighted

                if best is None or gain > best['gain']:
                    threshold = (col_sorted[i] + col_sorted[i+1]) / 2.0
                    best = {'feature': j, 'threshold': threshold, 'gain': gain}

        return best

    def build(self, X, y, depth):
        node = Node(depth)
        n_samples = len(y)
        node.n_samples = n_samples
        node.impurity = self.mse(y)

        stop = (
            node.impurity == 0 or
            n_samples < self.min_samples_split or
            (self.max_depth is not None and depth >= self.max_depth)
        )

        split: Dict | None = None
        if not stop:
            split = self.best_split(X, y)
            if split is None or split['gain'] < self.min_mse_decrease:
                stop = True

        if stop or split is None:
            node.value = np.mean(y)
            self.n_leaves += 1
            return node

        if self.feature_importances is not None:
            self.feature_importances[split['feature']] += (n_samples / self.n_total_samples) * split['gain']

        j, t = split['feature'], split['threshold']
        left = X[:, j] <= t

        node.feature = j
        node.threshold = t

        node.left = self.build(X[left], y[left], depth+1)
        node.right = self.build(X[~left], y[~left], depth+1)
        return node

    def prune(self, node):
        if node is None:
            return 0.0, 0

        if node.is_leaf:
            cost = (node.n_samples / self.n_total_samples) * node.impurity
            return cost, 1

        cost_left, leaves_left = self.prune(node.left)
        cost_right, leaves_right = self.prune(node.right)

        cost_children = cost_left + cost_right
        leaves_children = leaves_left + leaves_right

        cost_node = (node.n_samples / self.n_total_samples) * node.impurity

        if cost_node <= cost_children + self.ccp_alpha * (leaves_children - 1):
            node.left = None
            node.right = None
            self.n_leaves_ -= (leaves_children - 1)
            return cost_node, 1

        return cost_children, leaves_children

    def fit(self, X, y):
        X = np.asarray(X, dtype=float)
        y = np.asarray(y, dtype=float)
        n_samples, n_features = X.shape
        self.feature_importances = np.zeros(n_features)
        self.n_total_samples = n_samples
        self.n_leaves = 0
        self.root = self.build(X, y, 0)
        total = self.feature_importances.sum()
        if total > 0:
            self.feature_importances /= total
        return self

    def predict_one(self, x, node):
        while not node.is_leaf():
            node = node.left if x[node.feature] <= node.threshold else node.right
        return node.value

    def predict(self, X):
        X = np.asarray(X, dtype=float)
        return np.array([self.predict_one(x, self.root) for x in X])

    def depth(self):
        def d(node: Node|None):
            if node is None:
                return 0
            if node.is_leaf():
                return node.depth
            return max(d(node.left), d(node.right))
        return d(self.root)
````

## 4. Metric đánh giá
### Cây Hồi quy
- MSE (Mean Squared Error - Sai số toàn phương trung bình): Trung bình của bình phương các sai số. Vì tính bình phương, MSE trừng phạt rất nặng các điểm dữ liệu dự đoán sai lệch lớn (outliers).
  
$$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

- RMSE (Root Mean Squared Error - Căn bậc hai của MSE): Bằng cách lấy căn bậc hai, RMSE đưa sai số về cùng đơn vị với biến mục tiêu (target variable), giúp con số này dễ diễn giải hơn so với MSE.
  
$$RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$

- MAE (Mean Absolute Error - Sai số tuyệt đối trung bình): Trị tuyệt đối khoảng cách trung bình giữa dự đoán và thực tế. MAE ít bị ảnh hưởng bởi các giá trị ngoại lai (outliers) hơn so với MSE.
  
  $$MAE = \frac{1}{n} \sum_{i=1}^{n} \vert{}y_i - \hat{y}_i\vert{}$$
  
- $R^2$ Score (Hệ số xác định): Cho biết bao nhiêu phần trăm phương sai của biến phụ thuộc (thực tế) được giải thích bởi mô hình.
  $$R^2 = \frac{SS_{reg}}{SS_{total}} = 1 - \frac{SS_{res}}{SS_{total}}$$
  - $R^2 = 1$: Mô hình dự đoán hoàn hảo.
  - $R^2 = 0$: Mô hình chỉ dự đoán bằng giá trị trung bình của tập dữ liệu.
  - $R^2 < 0$: Mô hình dự đoán còn tệ hơn cả việc lấy trung bình.


### Cây phân loại
- Ma trận nhầm lẫn:
  
  

- Ma trận nhầm lẫn (Confusion Matrix)
    - TP (True Positive): Cây quyết định phân loại đúng lớp Positive (Thực tế Positive, Cây dự đoán Positive).
    - TN (True Negative): Cây quyết định phân loại đúng lớp Negative (Thực tế Negative, Cây dự đoán Negative).
    - FP (False Positive - Lỗi loại I): Cây quyết định phân loại nhầm thành lớp Positive (Thực tế Negative, nhưng Cây dự đoán Positive).
    - FN (False Negative - Lỗi loại II): Cây quyết định bỏ sót lớp Positive (Thực tế Positive, nhưng Cây dự đoán Negative).
- Độ chính xác (Accuracy): Tỉ lệ các mẫu dữ liệu được cây phân loại đưa vào đúng node lá tương ứng trên tổng số mẫu.
  $$Accuracy = \frac{TP + TN}{TP + TN + FP + FN}$$
- Độ chuẩn xác (Precision): Trong số tất cả các mẫu mà cây quyết định kết luận là Positive, có bao nhiêu mẫu thực sự đúng là Positive.
  $$Precision = \frac{TP}{TP + FP}$$
- Độ bao phủ (Recall/Sensitivity): Trong toàn bộ các mẫu thực tế là Positive, nhánh phân loại của cây đã tìm ra và phân loại đúng được bao nhiêu phần trăm.
  $$Recall = \frac{TP}{TP + FN}$$
- F1-Score: Trung bình điều hòa của Precision và Recall. Chỉ số này trừng phạt nặng mô hình cây quyết định nếu có sự chênh lệch lớn giữa Precision và Recall
  $$F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall}$$

````python
def acc(y_true, y_pred):
    return np.sum(y_true == y_pred) / len(y_true)

def prec_recall_f1_score(y_true, y_pred, n_classes):
    precs = []
    recalls = []
    f1_scores = []
    for c in range(n_classes):
        TP = np.sum((y_pred == c) & (y_true == c))
        FP = np.sum((y_pred == c) & (y_true != c))
        FN = np.sum((y_pred != c) & (y_true == c))
        prec = TP / (TP + FP)
        recall = TP / (TP + FN)
        f1_score = 2 * prec * recall / (prec + recall)
        precs.append(prec)
        recalls.append(recall)
        f1_scores.append(f1_score)
    return np.mean(precs), np.mean(recalls), np.mean(f1_scores)
````

# Random Forest
## 1. Mô hình
- Được xây dựng theo hướng **Bagging** (Bootstrap Aggregating). Hiểu đơn giản là xây dựng một **khu rừng** gồm nhiều **Cây quyết định** độc lập với nhau và tổng hợp kết quả của chúng để đưa ra kết quả cuối cùng.
  

## 2. Bài toán áp dụng
- **Classification**: dự đoán nhãn rời rạc (vote đa số từ các cây)
- **Regression**: dự đoán giá trị liên tục (trung bình dự đoán từ các cây)

## 3. Thuật toán xây dựng Random Forest
- $D$: Tập dữ liệu đầu vào của bài toán, $D$ có N hàng (mẫu) và có M cột (đặc trưng).
- $K$: số lượng cây trong rừng
- $m$: số lượng đặc trưng được chọn ngẫu nhiên tại mỗi lần phân tách của **Node**
    - $m = \sqrt{M}$ cho bài toán Phân loại
    - $m = \frac{M}{3}$ cho bài toán Hồi quy
- Thuật toán:
  - Lặp $K$ lần để xây dựng $K$ cây quyết định
  - Xây dựng tập dữ liệu $D_k$
      - Tại mỗi vòng lặp, bốc ngẫu nhiên $N$ lần từ tập gốc $D$ để cho vào $D_k$
      - Trong $D_k$ sẽ có những hàng trùng nhau, có khoảng 37% dữ liệu trong $D$ sẽ không có trong $D_k$
   - Khởi tạo cây $T_k$ từ tập dữ liệu $D_k$
   - Trả về tập hợp gồm $K$ cây quyết định $\{T_1, T_2, ..., T_k\}$

### Out-of-Bag (OOB) Score
- Tập dữ liệu **bootstrap** được bốc ngẫu nhiên có hoàn lại $N$ lần trong $N$ mẫu gốc. Xác suất để tập bootstrap đó không 1 mẫu cụ thể là:
  $$(1-\frac{1}{N})^N$$
- Khi $N$ đủ lớn thì xác suất này sẽ khoảng 37%
- Có nghĩa là sẽ có khoảng 37% số mẫu gốc sẽ không xuất hiện trong tập bootstap, tập này gọi là Out-of-bag (OOB)
- Chính vì tập OOB không có trong tập dữ liệu huấn luyện của cây, nên có thể dùng làm dữ liệu kiểm tra cho cây đó ma không cần chia tập validation như K-fold cross validation
- Với bài toán phân loại:
    * Với mỗi mẫu $x_i$, chỉ những cây nào **không** dùng $x_i$ để train mới được "bỏ phiếu" dự đoán cho $x_i$.
    * Nhãn dự đoán OOB của $x_i$ = nhãn được nhiều cây (trong số các cây coi $x_i$ là OOB) bỏ phiếu nhất.
  $$OOB\_score = \frac{1}{n}\sum_{i=1}^n \mathbb{1}\left[\hat{y}_i^{OOB} = y_i\right]$$
- Với bài toán hồi quy:  thay vì vote đa số ta lấy trung bình các dự đoán OOB, rồi tính  R2  giữa dự đoán OOB và giá trị thật.
  $$OOB\_Score = 1 - \frac{\sum_{i=1}^{N} (y_i - \hat{y}_i^{oob})^2}{\sum_{i=1}^{N} (y_i - \bar{y})^2}$$

### Cài đặt code:
**Bài toán Phân loại:**

````python
import numpy as np
from CARTClassifier import CARTClassifier

class RandomForestClassifier:
    def __init__(self, n_estimators = 100, max_depth = None, max_features = 'sqrt'):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.max_features = max_features
        self.trees = []
        self.feature_importances = None

    def fit(self, X, y):
        X = np.asarray(X, dtype=float)
        y = np.asarray(y, dtype=float)
        self.trees = []
        n_samples, n_features = X.shape
        self.n_classes = int(np.max(y)) + 1
        self.feature_importances = np.zeros(n_features)
        # số phiếu bầu mẫu i thuộc class c, tính gộp qua tất cả các cây
        oob_votes = np.zeros((n_samples, self.n_classes))
        # tổng số phiếu bầu của mỗi mẫu i (Số cây mà mẫu i thuộc tập OOB)
        oob_counts = np.zeros(n_samples)
        for _ in range(self.n_estimators):
            bootstrap_indices = np.random.choice(a=n_samples, size=n_samples, replace=True)

            X_bootstrap = X[bootstrap_indices]
            y_bootstrap = y[bootstrap_indices]

            tree = CARTClassifier(max_depth = self.max_depth, max_features = self.max_features)

            tree.fit(X_bootstrap, y_bootstrap)
            if tree.feature_importances is not None:
                self.feature_importances += tree.feature_importances

            self.trees.append(tree)

            in_bag = np.zeros(n_samples, dtype=bool)
            in_bag[bootstrap_indices] = True
            oob_idx = np.where(~in_bag)[0]
            if len(oob_idx) > 0:
                preds = tree.predict(X[oob_idx])
                for idx, p in zip(oob_idx, preds):
                    oob_votes[idx, p] += 1
                    oob_counts[idx] += 1

        self.feature_importances /= self.n_estimators

        has_oob = oob_counts > 0
        oob_pred = np.argmax(oob_votes[has_oob], axis=1)
        self.oob_score = np.mean(oob_pred == y[has_oob])
        self.oob_coverage = has_oob.mean()

    def predict(self, X):
        tree_predicts = np.array([tree.predict(X) for tree in self.trees]).T
        return np.array([np.argmax(np.bincount(pred)) for pred in tree_predicts])
````

**Bài toán Hồi quy**

````python
from CARTRegressor import CARTRegressor
import numpy as np

class RandomForestRegressor:
    def __init__(self, n_estimators = 100, max_depth = None, max_features = None):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.max_features = max_features
        self.trees = []
        self.feature_importances = None

    def fit(self, X, y):
        X = np.asarray(X, dtype=float)
        y = np.asarray(y, dtype=float)
        self.trees = []
        n_samples, n_features = X.shape
        self.feature_importances = np.zeros(n_features)
        oob_sums = np.zeros(n_samples)
        oob_counts = np.zeros(n_samples)

        for _ in range(self.n_estimators):
            bootstrap_indices = np.random.choice(a=n_samples, size=n_samples, replace=True)

            X_bootstrap = X[bootstrap_indices]
            y_bootstrap = y[bootstrap_indices]

            tree = CARTRegressor(max_depth = self.max_depth, max_features = self.max_features)

            tree.fit(X_bootstrap, y_bootstrap)
            if tree.feature_importances is not None:
                self.feature_importances += tree.feature_importances

            self.trees.append(tree)

            in_bag = np.zeros(n_samples, dtype=bool)
            in_bag[bootstrap_indices] = True
            oob_idx = np.where(~in_bag)[0]
            for idx, p in zip(oob_idx, tree.predict(X[oob_idx])):
                oob_sums[idx] += p
                oob_counts[idx] += 1

        self.feature_importances /= self.n_estimators
        has_oob = oob_counts > 0
        if len(has_oob) > 0:
            y_pred = oob_sums[has_oob] / oob_counts[has_oob]
            y_true = y[has_oob]
            ss_total = np.sum((y_true - y_true.mean()) ** 2)
            ss_res = np.sum((y_true - y_pred) ** 2)
            self.oob_score = 1 - ss_res / ss_total

    def predict(self, X):
        tree_predicts = np.array([tree.predict(X) for tree in self.trees]).T
        return np.mean(tree_predicts, axis=1)
````
